```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
raimbow = cv2.imread('../DATA/rainbow.jpg1')
show_raimbow = cv2.cvtColor(raimbow, cv2.COLOR_BGR2RGB)
```

```python
img = raimbow
type(img)
```

```python
img.shape
```

```python
mask = np.zeros(img.shape[:2], np.uint8)
```

```python
mask[300:400, 100:400] = 255
```

```python
plt.imshow(mask, cmap='gray')
```

```python
mask_img = cv2.bitwise_and(img, img, mask = mask)
```

```python
show_masked_img = cv2.bitwise_and(show_raimbow, show_raimbow, mask= mask)
```

```python
plt.imshow(show_masked_img)
```

```python
hist_mask_values_red = cv2.calcHist([raimbow], channels=[2], mask=mask, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_mask_values_red)
plt.title('RED HISTOGRAM FOR MASKED RAINBOW') 
```

```python
hist_mask_values_red = cv2.calcHist([raimbow], channels=[2], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_mask_values_red)
plt.title('RED HISTOGRAM FOR RAINBOW')
```
