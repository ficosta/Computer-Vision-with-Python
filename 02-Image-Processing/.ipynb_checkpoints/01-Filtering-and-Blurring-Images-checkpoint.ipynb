# Filtering and Blurring Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/00-puppy.jpg')
```

```python
plt.imshow(img)
```

### Converting to Different Colorspaces

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

**Converting to HSV**
https://en.wikipedia.org/wiki/HSL_and_HSV

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img)
```

```python
img = cv2.imread('../DATA/00-puppy.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img)
```

# Filters and Effects

There are a lot of different effects and filters we can apply to images. We're just going to go through a few of them here. Most of them involve some sort of math based function being applied to all the pixels values

```python
def load_img():
    img = cv2.imread('../DATA/00-puppy.jpg').astype(np.float32) / 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
i = load_img()
plt.imshow(i)
```

### Gamma Correction : Practical Effect of Increasing Brightness

```python
img = load_img()
gamma = 1/4
effected_image = np.power(img, gamma)
plt.imshow(effected_image)
```

```python
img = load_img()
gamma = 4
effected_image = np.power(img, gamma)
plt.imshow(effected_image)
```
